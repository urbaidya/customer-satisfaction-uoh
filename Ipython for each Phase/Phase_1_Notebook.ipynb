### Remove the ID and seperate dataset to X_raw and y_raw

```python
column_names = raw_df.columns
print("These are the different column names we have:")
print(column_names)
print("-"*80)
print("We can maybe remove the columns 'ID' and 'TARGET', and store rest data in X_raw DataFrame and 'TARGET' in y_raw")
X_raw=raw_df.drop(["ID","TARGET"], axis=1)
y_raw=raw_df.TARGET
```

### Finding and Removing the constant value columns

```python
column_names = X_raw.columns
distinct_vals = []
for col in column_names:  ##here we are looping over all columns in the train data
    distinct_vals.append(len(X_raw[col].value_counts()))  ##in distinct_vals we are storing the distinct values for that feature

cols_data = pd.DataFrame({"Column_name":column_names, "Num_of_distinct_values":distinct_vals})

print(cols_data[cols_data.Num_of_distinct_values==1])  ##printing names of those columns which only have single value
                                                       ##in our case its mostly only 0 for these columns

print("-"*60)
print("CONCLUSION: We can see that many columns that are present have ONLY SINGLE VALUE and I think these columns can be dropped")
print("As these do not add any value to your model as they have only single value")
print("-"*60)
## store column names where we only have single unique values
const_value_cols = list(cols_data[cols_data.Num_of_distinct_values==1].Column_name)
print(f"Number of columns with only a single unique value ==> {len(const_value_cols)}")  ## there are 34 such columns
print("-"*60)
X_raw_1=X_raw.drop(const_value_cols, axis=1)  ## drop these columns from our train data
print("New shape of data after remove these columns")
print(X_raw_1.shape)  ##new shape after dropping these columns
```

### Finding and remove features which have identical values

```python
new_cols=X_raw_1.columns
same_vals=[]
same_value_pairs=[]
for i in new_cols:  ## loop over columns
    for j in new_cols:  ## loop over columns
        if i!=j:  ## proceed only if the name of the two columns are not same
            if (i not in same_vals) and (j not in same_vals):  ## check if the columns are not already processed
                flag=0  ## set indicator flag to 0
                for t1,t2 in zip(X_raw_1[i],X_raw_1[j]):  ## loop of values of both columns one by one
                    if t1!=t2:  ## check if at any time value is different for any datapoint then exit
                        flag=1  ## set indicator to 1 and break the loop
                        break
                if flag==0:  ## if indicator is 0 means column are same
                    same_value_pairs.append((i,j))  ## store the columns which have same values
                    same_vals.append(i)  #store only a feature one
```

As these columns have same values keeping only one of these is sufficient and this can also be checked using correlation

```python
from scipy.stats import pearsonr,spearmanr  ##import libraries to get spearman and pearson correlation coefficient
for var1,var2 in same_value_pairs:
    pcorr, _ = pearsonr(X_raw_1[var1],X_raw_1[var2])  ##calculate the ppc
    scorr, _ = spearmanr(X_raw_1[var1],X_raw_1[var2])  ##calculate the scc
    print(f"For the {var1} and {var2} the value of pcc is {round(pcorr,10)} and scc is {round(scorr,10)}")
```

Here we can remove any one value of these pairs as the information store in anyone of these variable is completely available in the second variable in that pair as these features have exactly same value hence they are highly correlated.

```python
X_raw_2=X_raw_1.drop(same_vals, axis=1)  ## drop these columns from our train data
print("New shape of data after remove these columns")
print(X_raw_2.shape)  ##new shape after dropping these columns
```

### Code for Precision-Recall Curve and AUC

```python
def getConfusionMatrix(data):
    """
    This function calculates and return values for:
    1. True Positive
    2. False Negative
    3. False Positive
    4. True Negative
    """
    TP = len(data[(data.y == 1) & (data.y_hat == 1)].index)
    FN = len(data[(data.y == 1) & (data.y_hat == 0)].index)
    FP = len(data[(data.y == 0) & (data.y_hat == 1)].index)
    TN = len(data[(data.y == 0) & (data.y_hat == 0)].index)
    return(TP, FN, FP, TN)


def calcTPRnFPR(data, thres):
    """
    This function use the threshold value to get the predicted value and return the 
    Precision and Recall for that particular threshold and the given dataset.
    """
    data['y_hat'] = list(map(lambda x: 0 if x < thres else 1, data.proba))
    TP, FN, FP, TN = getConfusionMatrix(data)
    Precision = TP / (TP + FP)
    Recall = TP / (TP + FN)
    return(Precision, Recall)


def calcMetrics(data):
    """
    This function calculates the value for 
    1. AUC Score &
    2. Plot PR curve
    """
    Precisions = []
    Recalls = []    
    TP, FN, FP, TN = getConfusionMatrix(data)
    
    thresholds = np.array(sorted(set(data.proba), reverse = True))  # get unique values of the threshold

    for threshold in thresholds:
        Precision , Recall = calcTPRnFPR(data[['y', 'proba']], threshold)
        Precisions.append(Precision)
        Recalls.append(Recall) 
    
    PR_auc_score = np.trapz(Precisions, Recalls)  ## https://stackoverflow.com/questions/53603376/calculate-auc-manually-without-using-any-numpy-or-sklearn-library
    plot_auc(Precisions, Recalls)

    
def plot_PRc(Precisions, Recalls):
    """
    This function plots the PR Curve.
    """
    import matplotlib.pyplot as plt
    plt.title('Precision-Recall Curve')
    plt.xlabel('Recalls')
    plt.ylabel('Precisions')
    plt.plot(Recalls, Precisions)
    plt.plot([0,1], [0,1], 'k--')
    plt.show()
```
